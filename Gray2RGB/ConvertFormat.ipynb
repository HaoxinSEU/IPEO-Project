```python
from PIL import Image
import numpy as np
import os
```

```python
folder_path = "./Forest_gray/input/train"

# Get a list of all files in the folder
all_files = os.listdir(folder_path)

# Filter grayscale images
grayscale_images = [file for file in all_files if Image.open(os.path.join(folder_path, file)).mode == 'L']

# Print the list of grayscale images
print("Grayscale Images:")
print(grayscale_images)
```

```python
# Iterate through grayscale images and convert them to RGB
for file in grayscale_images:
    file_path = os.path.join(folder_path, file)

    # Open the image
    img = Image.open(file_path)
    
    # convert to RGB
    img = img.convert('RGB')

    # remove original grayscale images
    os.remove(file_path)
    
    # Save the converted image
    img.save(file_path)
```
